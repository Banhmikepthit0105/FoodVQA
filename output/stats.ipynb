## Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load BLEU data
llama_bleu = pd.read_csv("llama_bleu_results.csv")
miniCPM_bleu = pd.read_csv("miniCPM_bleu_results.csv")
qwen_bleu = pd.read_csv("qwen_bleu_results.csv")

# Load ROUGE data
llama_rouge = pd.read_csv("llama_rouge_results.csv")
miniCPM_rouge = pd.read_csv("miniCPM_rouge_results.csv")
qwen_rouge = pd.read_csv("qwen_rouge_results.csv")

# Add a 'model' column to each dataframe for identification
llama_bleu['model'] = 'LLaMA'
miniCPM_bleu['model'] = 'MiniCPM'
qwen_bleu['model'] = 'Qwen'

llama_rouge['model'] = 'LLaMA'
miniCPM_rouge['model'] = 'MiniCPM'
qwen_rouge['model'] = 'Qwen'

# Combine BLEU data into one dataframe
bleu_combined = pd.concat([llama_bleu, miniCPM_bleu, qwen_bleu], ignore_index=True)

# Combine ROUGE data into one dataframe
rouge_combined = pd.concat([llama_rouge, miniCPM_rouge, qwen_rouge], ignore_index=True)

# 1. Bar Plot: Average BLEU Score per Model
plt.figure(figsize=(10, 6))
sns.barplot(x='model', y='bleu_score', data=bleu_combined, ci=None)
plt.title('Average BLEU Score per Model')
plt.xlabel('Model')
plt.ylabel('Average BLEU Score')
plt.show()

# 2. Box Plot: BLEU Score Distribution per Model
plt.figure(figsize=(10, 6))
sns.boxplot(x='model', y='bleu_score', data=bleu_combined)
plt.title('BLEU Score Distribution per Model')
plt.xlabel('Model')
plt.ylabel('BLEU Score')
plt.show()

# 3. Bar Plot: Average ROUGE Scores per Model
# Melt the ROUGE dataframe to plot multiple ROUGE types
rouge_melted = rouge_combined.melt(id_vars=['model'], value_vars=['rouge1', 'rouge2', 'rougeL'],
                                   var_name='rouge_type', value_name='score')

plt.figure(figsize=(12, 6))
sns.barplot(x='model', y='score', hue='rouge_type', data=rouge_melted, ci=None)
plt.title('Average ROUGE Scores per Model')
plt.xlabel('Model')
plt.ylabel('Average ROUGE Score')
plt.legend(title='ROUGE Type')
plt.show()

# 4. Violin Plot: BLEU Score Distribution
plt.figure(figsize=(10, 6))
sns.violinplot(x='model', y='bleu_score', data=bleu_combined)
plt.title('BLEU Score Distribution per Model (Violin Plot)')
plt.xlabel('Model')
plt.ylabel('BLEU Score')
plt.show()

# 5. Box Plot: ROUGE-L Distribution (example for one ROUGE type)
plt.figure(figsize=(10, 6))
sns.boxplot(x='model', y='rougeL', data=rouge_combined)
plt.title('ROUGE-L Score Distribution per Model')
plt.xlabel('Model')
plt.ylabel('ROUGE-L Score')
plt.show()
```
